# read the dataset

```python
sports_dt = pd.read_csv('sports.csv')
sports_dt
```

The dataset is a collection of events in a football/soccer match where each row corresponds to events that happened to a single match. Each row is identified by its unique identifier for a snapshot and each unique identifier also contains on what minute of the match the snapshot occured.

Now let's see what the columns are for.

```python
sports_dt.head
```

There are 21 columns in the dataset and these 21 columns are
-  uuid                    : Identifier
-  current_minute          : current minute the snapshot was taken. NOTE: minutes include 15 min breaktime in the halftime
-  home_score              : current score of the home team by the current_minute
-  away_score              : current score of the away team by the current_minute
-  home_yellow_cards       : number of yellow cards the home team incurred by the current_minute
-  away_yellow_cards       : number of yellow cards the away team incurred by the current_minute
-  home_red_cards          : number of red cards the home team incurred by the current_minute
-  away_red_cards          : number of red cards the away team incurred by the current_minute
-  home_attacks            : attacks attempted by the home team by the current_minute
-  away_attacks            : attacks attempted by the away team by the current_minute
-  home_dangerous_attacks  : Dangerous attacks attempted by the home team by the current_minute
-  away_dangerous_attacks  : Dangerous attacks attempted by the away team by the current_minute
-  home_corners            : Corner kicks awarded to the home team by the current_minute
-  away_corners            : Corner kicks awarded to the away team by the current_minute
-  home_off_target         : goal attempts by the home team that did not need to be defended by the goalkeeper
-  away_off_target         : goal attempts by the away team that did not need to be defended by the goalkeeper
-  home_on_target          : goal attempts by the home team that needs to be defended by the goalkeeper
-  away_on_target          : goal attempts by the away team that needs to be defended by the goalkeeper
-  home_possession
-  away_possession
